# League of legend data analysis
by Arthur Berthelier le 29 avril 2021

**Description :**

Le jeux de donnée choisi est issu du célébre jeux en ligne "free to play", league of legend sortie en 2019, le célèbre MOBA de Riot Games (multiplayer online battle arena) est sûrement le jeu le plus suivi au niveau "esport", mais également l’un des jeux multijoueur les plus joués au monde, et ce depuis une dizaine d’années déjà.

**Objectif du jeux :**

  Dans une partie deux équipes de 5 joueurs s'affronte et on pour objectif, la destruction du "nexus" représentant l'élement principal de la base ennemie. Pour arriver a ce résultat plusieurs facteurs son à prendre en compte comme les ojectifs et batiments a détruire,le choix du personnage que l'on banni et celui que l'on choisi,le choix de deux sorts d'invocateur, la durée de la partie...
  lol.png

**Sélection du personnage :**

Au début de la partie chacun des joueurs choisit un personnage a bannir, qui ne pourra donc pas être choisit par les deux équipes, ensuite chacun des joueurs est invités à choisir un personnage parmis la liste des personnages disponnibles (150 - les personnages bannie) et deux sort d'invocateur parmis les 8 proposés, les champions son unique mais plusieurs joueurs peuvent choisir un même sort d'invocateur.

**Les Données annalisées :**

Nous avons en notre possésion un échantillion conséquent de 50 mille partie en mode classé ( mode compétitif ) qui se sont déroulées pendant la saison 9 de league of Legends, donc du 23 janvier au 19 novembre 2019. Chaque partie comporte différentes données pouvant influencer le dénoument de celle-ci, comme sa durée, sa date de création, le choix des personnage des deux équipes, le choix des personnages bannis et différente information sur les premiers objectifs prit et leurs nombres par équipe. De plus cette échantillion est issue du serveur d'Europe de l'ouest et ne comporte pas le poste ni l'équipement du personnage du joueur afin de le rendre plus représentatif. 
En effet l'équipement et le poste du joueur sont des facteurs important mais peuvent fausser notre résultat car dans l'hypothése ou un joueur chosit un personnage avec 100% de chance de victoire, si celui-ci décide de jouer ce personnage a un poste différents du poste habituel du personnage avec un équipement différents de l'équipement habituel du personnage il pourais ce retouver facilement avec 0% de victoire.


**Problématique**

J'ai pour objectif de réaliser une étude statistique et graphique sur les meilleurs conditions de victoire en millieux compétitif lors de la saison 9 de league of legends, cela dans le but de pouvoir identifier la combinaison de facteur garantissant le plus fort pourcentage de victoire et ainsi pouvoir réaliser de la prédiction sur le résultats d'une partie en saison 9. 

**Amélioration de l'étude**

En partant des résultat de cette étude on pourais aussi l'appliquer par la suite aux différentes saison de league of legends depuis le lancement du mode compétitif et comparer les résultats de nos études, et ainssi pouvoir évaluer l'évolution de la statégie d'équilibrage du jeux de l'entreprise Riot Game et pouvoir prévoir des compositions gagnantes pour les saisons a venir, ce qui serais trés utiles et démandé par les équipes E-sport league of legend du monde entier.
